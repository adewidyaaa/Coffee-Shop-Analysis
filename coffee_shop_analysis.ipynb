## v. Exploration & Analysis

Bagian ini berisikan analisis data yang mencakup visualisasi data, perhitungan statistik deskriptif dan statistik inferensial. Fokus analisa meliputi problem statement yang sudah dijabarkan sebelumnya. 1. Menu kopi apa yang paling sering dibeli oleh pelanggan ?  
Analisa akan dilakukan dengan menghitung persentase jumlah pemesanan untuk tiap menu kopi dari data transaksi. Dan hasilnya akan divisualisasikan menggunakan Pie Chart.

```python
df.groupby('coffee_name')['coffee_name'].count().sort_values(ascending=False)\
.plot(kind='pie', 
        autopct = '%.2f%%', 
        figsize=(8,8), 
        ylabel="",
        title= "Menu Kopi Terlaris" )
'''
Menampilkan total persentase pembelian pada kolom coffee_name dengan menggunakan pie chart.
'''
plt.show()
```

Dari data yang ditampilkan pada chart diatas, disimpulkan bahwa Americano with Milk dan Latte merupakan menu kopi terlaris yang ada di kedai kopi dengan persentase masing-masing sebesar 22.81% dan 21.34% dari total transaksi yang ada. Kemudian, menu yang cukup populer juga adalah Americano dan Cappucino dengan persentase 15.90% dan 13.70%, sedangkan untuk menu yang lain termasuk kurang diminati dikarenakan hanya memiliki persentase penjualan > 10%. 
Berdasarkan hal ini. manajemen kedai dapat melakukan pengelolaan stok dan promosi lebih berfokus pada menu yang memiliki persentase tertinggi, sementara itu untuk menu yang kurang diminati dapat dilakukan pengembangan kembali.

2. Menu kopi apa yang memberikan total omset paling tinggi ?  
Analisa akan dilakukan dengan menghitung jumlah omset untuk tiap menu kopi dari data transaksi. Dan hasilnya akan divisualisasikan menggunakan Horizontal Bar Chart.

```python
df.groupby('coffee_name')['money'].sum().sort_values()\
.plot(kind= 'barh',
      figsize= (12,6),
      xlabel= "Total Omset",
      ylabel= "Menu Kopi",
      title= "Total Omset per Menu")
'''
Menampilkan total omset dari tiap menu kopi dalam 1 tahun terakhir,
diurutkan berdasarkan omset terbesar.
'''

print(df.groupby('coffee_name')['money'].sum().sort_values())

plt.show()
```

Dari data diatas dapat dilihat bahwa Latte merupakan menu di kedai dengan total omset tertinggi sepanjang tahun yaitu 26875, diikuti oleh Americano with Milk dengan total omset sebesar 24751. Sedangkan Espresso merupakan menu dengan total omset terendah sepanjang tahun yaitu sebesar 2690. Dapat disimpulkan bahwa pelanggan lebih menyukai kopi dengan basis susu dibandingkan dengan kopi hitam ataupun minuman cokelat lainnya.

3. Pada jam berapa terjadinya peak hour di kedai kopi ?  
Analisa akan dilakukan dengan menghitung jumlah transaksi yang terjadi tiap jamnya dan mencari jam dengan jumlah transaksi tertinggi. Kemudian hasilnya akan divisualisasikan menggunakan Line Chart.

```python
df['hour'] = df['time'].dt.hour

df.groupby('hour')['money'].count()\
.plot(figsize= (12,6),
      marker= 's',
      grid= True,
      title= "Peak Hour Kedai Kopi",
      xlabel="Jam",
      ylabel= "Jumlah Transaksi")

plt.xticks(range(6,23))

print(df.groupby('hour')['money'].count())

plt.show()
```

Dari data diatas dapat dilihat bahwa peak hour di kedai kopi terjadi pada pukul 10.00 pagi dengan jumlah transaksi tertinggi yang terjadi sepanjang tahun sebesar 328 transaksi. Selain itu, pada pukul 11.00 dan 16.00 juga termasuk periode dengan transaksi yang cukup ramai di kedai kopi. Secara umum, pagi dan sore hari merupakan kategori waktu dengan aktivitas transaksi tertinggi yang terjadi di kedai.

4.  Hari dengan total omset tertinggi dalam seminggu ?  
Analisa akan dilakukan dengan menjumlahkan semua transaksi per hari berdasarkan kategori weekday. Memberikan gambaran distribusi omset per hari dalam seminggu.Kemudian hasilnya akan divisualisasikan menggunakan Bar Chart.

```python
hari = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

df['weekday'] = pd.Categorical(df['weekday'], categories=hari, ordered=True)

df.groupby('weekday')['money'].sum()\
.plot(kind='bar',
    figsize=(10,6),
    title="Hari dengan Total Omset Tertinggi",
    ylabel="Total Omset",
    xlabel="Weekday")

print(df.groupby('weekday')['money'].sum())

plt.show()
```

Dari chart diatas dapat disimpulkan bahwa kedai kopi cenderung memiliki omset lebih tinggi pada hari kerja (weekdays) dibandingkan akhir pekan (weekends).
Hal ini dapat dilihat dari total omset tertinggi didapatkan pada hari Senin dan Selasa, sedangkan untuk hari Sabtu dan Minggu merupakan hari dengan total omset terendah di sepanjang minggu.

5. Berapa rata-rata jumlah transaksi harian tiap bulan selama 1 tahun terakhir ?  
Analisa akan dilakukan dengan menghitung rata-rata transaksi harian untuk tiap bulan. Kemudian dilakukan penjabaran dengan statistik deskriptif menggunakan perhitungan central tendency (mean, median, mode).

```python
# Mengambil nilai bulan dalam kolom date
df['month'] = df['date'].dt.strftime('%b')

# Menghitung transaksi per hari dalam bulan
trans_harian = df.groupby(['month', 'date'])['money'].count().reset_index(name='jml_trans')

# Menghitung central tendency per bulan
hasil = trans_harian.groupby('month')['jml_trans'].agg(['mean', 'median', pd.Series.mode]).reset_index()

# Menampilkan hasil berdasarkan urutan bulan
bulan = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
hasil['month'] = pd.Categorical(hasil['month'], categories=bulan, ordered=True)
hasil = hasil.sort_values('month').reset_index(drop=True)

print(hasil)
```

Dari data diatas, terlihat bahwa jumlah transaksi harian tiap bulan cukup bervariasi dan tidak stabil. 
Bulan februari merupakan bulan dengan rata-rata jumlah transaksi harian tertinggi yaitu 15.10 transaksi, sedangkan bulan april memiliki rata-rata harian terendah yaitu 5.6 transaksi.
Selain itu, beberapa bulan ditengah tahun menunjukkan rata-rata jumlah transaksi lebih rendah dibandingkan diawal dan diakhir tahun.

6. Bagaimana tren revenue kedai kopi per bulan selama 1 tahun terakhir ?  
Akan dilakukan perhitungan dengan mencari total revenue per bulan lalu dibandingkan kenaikan atau penurunannya. Kemudian dilakukan penjabaran statistik inferensial menggunakan ANOVA.

```python
from scipy.stats import f_oneway

bulan = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

revenue_per_bulan = [df['money'][df['month_name']==m] for m in bulan]


f_stat, p_value = f_oneway(*revenue_per_bulan)  

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

Dari hasil perhitungan statistik inferensial menggunakan metode ANOVA diatas, dpat disimpulkan bahwa ada perbedaan revenue antar bulan signifikan. Beberapa bulan menunjukkan kenaikan revenue yang lebih tinggi dibanding bulan lain, menandakan adanya pola musiman atau fluktuasi penjualan bulanan.
